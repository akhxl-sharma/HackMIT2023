```python
# Define the folder containing the .csv files
csv_folder = '/Users/akhil/Downloads/HackMIT2023/EEG Stress Dataset/Trier Mental Challenge Test (TMCT) CSV'

# Define the output folder for the .pkl files
pkl_folder = '/Users/akhil/Downloads/HackMIT2023/EEG Stress Dataset/Trier Mental Test PKL'

csv_files = [f for f in os.listdir(csv_folder) if f.endswith('.csv')]

# Loop through each .csv file and convert it to .pkl
for csv_file in csv_files:
    csv_path = os.path.join(csv_folder, csv_file)
    
    # Read the .csv file using pandas
    df = pd.read_csv(csv_path)
    
    # Create the output .pkl file path
    pkl_file = os.path.splitext(csv_file)[0] + '.pkl'
    pkl_path = os.path.join(pkl_folder, pkl_file)
    
    # Save the DataFrame as a .pkl file using pickle
    with open(pkl_path, 'wb') as pkl_file:
        pickle.dump(df, pkl_file)

print("Conversion completed for", len(csv_files), "files.")
```
